```python


import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
#we will be creating a linear regression model for giving the probable no of marks for the hours studied
x=[[1],[2],[3],[4],[5]]#no of hours studied
y=[2,4,6,8,10]#marks achieved out of 10
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=20)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
model=LinearRegression()
model.fit(X_train,Y_train)
Y_pred=model.predict(X_test)
print(Y_pred)
print(Y_test)
correlation = np.corrcoef(Y_test, Y_pred)[0, 1]
print("Correlation between actual and predicted values:", correlation)  #correlation gives us a idea between to values i mean if they are increasing decreasin or what
r2=r2_score(Y_pred,Y_test)
print(f'the r2 of the model is {r2}')
mse=mean_squared_error(Y_pred,Y_test)
print(f'the mean square error is {mse}')


plt.scatter(X_test,Y_test,color="red")
plt.plot(X_test,Y_pred,color="blue")
plt.title("hours vs marks")
plt.xlabel("hours")
plt.ylabel("marks")
```
